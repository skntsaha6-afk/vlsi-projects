Let's create something which will ask question and answer the question.             

```python
question = "Bring out list of agentic ai oppurtunity 5-10 in the businness field"
message= [{

        "role" : "user",
        "content" : question

}]

```

```python
ans1 = openai.chat.completions.create(model="gpt-4.1-nano", messages=message)
print(ans1)
ans1 = ans1.choices[0].message.content  + "Bring out top 3 idea which can be breakthrough or can be worth exploring for commerial purpose"
print(ans1)
message= [{

    "role" : "user",
    "content" : ans1
}]
```

```python
final_rep = openai.chat.completions.create(model="gpt-4.1-mini", messages=message)
```

```python
print(final_rep.choices[0].message.content)
```

```python
question = " Consider yourself as a evaluator and tell me which one to focus on for the future proof, breackthrough development, how to proceed towards it"
final2_rep = final_rep.choices[0].message.content + '\n' + question
print(final2_rep)
message= [{
        "role" : "system",
        "content" : final2_rep
}]
res = openai.chat.completions.create(model="gpt-4.1-mini", messages=message)
```

```python
print(res.choices[0].message.content)
```
